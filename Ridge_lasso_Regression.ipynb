DATA CLEANING

```python
# Missing Values

dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:"Region"]=1
df=dataset
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
#Removing the Null Values
df=df.dropna().reset_index(drop=True)
```

```python
#remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
##remove the 122nd row
df=df.drop(122).reset_index(drop=True)
```

```python
## Fix spaces in column names
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
print(df[['day','month','year','Temperature','RH','Ws']].dtypes)
```

```python
cols_to_check = ['day','month','year','Temperature','RH','Ws']
for col in cols_to_check:
    print(f"\nUnique values in '{col}':")
    print(df[col].unique())
```

```python
for col in ['day','month','year','Temperature','RH','Ws']:
    df[col] = pd.to_numeric(df[col], errors='coerce')  # Converts invalid entries to NaN

# Optional: drop rows with NaN (if appropriate)
df.dropna(subset=['day','month','year','Temperature','RH','Ws'], inplace=True)

```

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.head()
```

```python
df.info()
```

CHANGING THE OTHER COLUMNS TO FLOAT DATA DATATYPE

```python
objects=[features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
#drop day month and year
df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
#categories in classes
df_copy['Classes'].value_counts()
```

```python
#Encoding of the categories in classes
df_copy['Classes'] = np.where(df_copy['Classes'].astype(str).str.contains('not fire', case=False), 0, 1)
```

```python
#percentage for pie chart
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
#Plotting piechart
classlabels=["Fire","Not Fire"]
percentage = [60, 40]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```
